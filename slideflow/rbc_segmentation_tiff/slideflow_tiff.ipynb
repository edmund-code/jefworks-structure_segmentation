```python
#create slideflow project

import slideflow as sf
import slideflow.util as sf_util


setattr(sf, 'util', sf_util)

P = sf.create_project(
  root='/Users/edmundtsou/Desktop/JEFworks/jefworks-structure_segmentation/rbc_segmentation',
  annotations="/Users/edmundtsou/Desktop/JEFworks/jefworks-structure_segmentation/rbc_segmentation/annotations.csv",
  slides='/Users/edmundtsou/Desktop/JEFworks/jefworks-structure_segmentation/data/lab_kidney_data_tiff',
)
```

```python
P.extract_tiles(
    tile_px=512,
    tile_um=256,   # This equals 0.5 MPP (512px * 0.5 microns/px = 256um)
    mpp_override=0.5,       # Manually set resolution for .tif files
    save_tiles=True
)
```

```python
import slideflow as sf
from slideflow import cellseg
import matplotlib.pyplot as plt
import numpy as np

# 1. Get the slide path using the dataset object
all_slides = P.dataset().slides()
if not all_slides:
    raise ValueError("No slides found in project.")

slide_name = all_slides[0]
# Use the dataset object to resolve the full path
slide_path = P.dataset().path(slide_name) 
print(f"Testing on: {slide_path}")

# 2. Extract and segment a patch
# Level 0 is the highest resolution (usually 20x or 40x)
slide = sf.WSI(slide_path, backend='libvips')
patch = slide.read_region(location=(5000, 5000), level=0, size=(1024, 1024))

# 3. Run segmentation on the extracted patch
# This will use the backend you have installed (torch/mac GPU)
mask = cellseg.segment_image(
    patch,
    model='cellpose', 
    diam_um=8,
    backend='torch'
)

# 4. Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].imshow(patch)
axes[0].set_title(f"Original Patch: {slide_name}")
axes[0].axis('off')

axes[1].imshow(mask, cmap='nipy_spectral')
axes[1].set_title("RBC Segmentation Mask")
axes[1].axis('off')

plt.tight_layout
```
